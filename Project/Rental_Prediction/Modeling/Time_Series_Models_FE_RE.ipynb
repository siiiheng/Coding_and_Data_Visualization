```python
df = pd.read_csv("data_final_with_scaled.csv", index_col=0)
df.head()
```

## Data Preprocessing

```python
df1 = df.iloc[:, 0:3]
df1
```

```python
df2=df.iloc[:,3:]
df2['tract1']=df['tract']
df2
```

```python
df2.head()
```

```python
data = df_final
data
```

```python
#Set for 2 layer index
data = data.set_index(["tract", "year"])
```

```python
data.head(10)
```

```python
df = data
df.head()
```

## Fixed Effect Model

```python
#Tuning for alpha
y_var_name = 'mean_rent_per'
X_var_names = ['total_units', 'property_sqft',
       'current_rent', 'property_value', 'income_tax_rate',
       'corporate_tax_rate', 'state_sales_tax_rate',
       'avg_local_sales_tax_rate', 'combined_rate', 'max_local_sales_tax',
       'median_income', 'population',
       'white_pop', 'black_pop', 'asian_pop', 'pacific_pop', 'pop_0_19', 'pop_20_29', 'pop_30_39', 'pop_40_49',
       'pop_50_59', 'pop_60_69', 'pop_70_over', 'occupancy_rate',
       'house_supply', 'crime_rate', 'ROI_SP500', 'ROI_Tbill', 'ROI_Tbond',
       'corp_bond', 'real_estate', 'sp500', 'Tbill', 'Tbond5', 'baa_corp_bond',
       'real_estate2', 'stocks_bills', 'stocks_bonds', 'stocks_corp_bonds',
       'hist_risk', 'inflation', 'SP500', 'Tbill_3month', 'Tbond_10year',
       'baa_corp_bonds', 'real_estate3',
       'grocery_vists_per', 'hospitals_per', 'colleges_per',
       'private_schools_per', 'employment_rate']
```

```python
exog_vars = X_var_names
exog = sm.add_constant(df[exog_vars])
mod = PanelOLS(df.mean_rent_per, exog, entity_effects = True, check_rank=False, drop_absorbed=True)
FE = mod.fit()
print(FE)
```

## Random Effect Model with Same Varaibles as Fixed Effect Model


```python
#Tuning for alpha
y_var_name = 'mean_rent_per'
X_var_names = ['total_units', 'property_sqft',
       'current_rent', 'property_value', 'income_tax_rate',
       'corporate_tax_rate', 'state_sales_tax_rate',
       'avg_local_sales_tax_rate', 'combined_rate', 'max_local_sales_tax',
       'median_income', 'population',
       'white_pop', 'black_pop', 'asian_pop', 'pacific_pop', 'pop_0_19', 'pop_20_29', 'pop_30_39', 'pop_40_49',
       'pop_50_59', 'pop_60_69', 'pop_70_over', 'occupancy_rate',
       'house_supply', 'crime_rate', 'ROI_SP500', 'ROI_Tbill', 'ROI_Tbond',
       'corp_bond', 'real_estate', 'sp500', 'Tbill', 'Tbond5', 'baa_corp_bond',
       'real_estate2', 'stocks_bills', 'stocks_bonds', 'stocks_corp_bonds',
       'hist_risk', 'inflation', 'SP500', 'Tbill_3month', 'Tbond_10year',
       'baa_corp_bonds', 'real_estate3',
       'grocery_vists_per', 'hospitals_per', 'colleges_per',
       'private_schools_per', 'employment_rate']
```

```python
exog_vars = X_var_names
exog = sm.add_constant(df[exog_vars])
mod = PanelOLS(df.mean_rent_per, exog, entity_effects = False, check_rank=False, drop_absorbed=True)
RE = mod.fit()
print(RE)
```

## Fixed Effect Model Prediction

```python
df_1719 = df.loc[(df.index.get_level_values('year') <= 2019)]
df_19 = df.loc[(df.index.get_level_values('year') == 2019)]
df_2022 = df.loc[(df.index.get_level_values('year') >= 2020)]
df_22 = df.loc[(df.index.get_level_values('year') == 2022)]
```

```python
para = pd.DataFrame(FE.params)
```

```python
df_22['const'] = float(para.loc['const'])
```

```python
new_x_var = ['const','total_units', 'property_sqft',
       'current_rent', 'property_value',
       'median_income', 'population',
       'white_pop', 'black_pop', 'asian_pop', 'pacific_pop', 'pop_0_19', 'pop_20_29', 'pop_30_39', 'pop_40_49',
       'pop_50_59', 'pop_60_69', 'pop_70_over', 'occupancy_rate',
       'house_supply', 'crime_rate', 'ROI_SP500', 'ROI_Tbill', 'ROI_Tbond',
       'grocery_vists_per', 'hospitals_per', 'colleges_per',
       'private_schools_per', 'employment_rate']

exog = df_22[new_x_var]
df_22['predict']= FE.predict(exog)
df_22['growth_25'] = df_22['predict']/df_22['mean_rent_per'] - 1
df_22.sort_values('growth_25', ascending = False) 
```

```python
FE_data_needed = df_22.sort_values('growth_25', ascending = False).head(10)
```

```python
FE_data_needed
```

```python
df_22 = df_22.reset_index( level = [0 , 1] )
```

```python
FE_data_needed =df_22.loc[:, ['tract', 'growth_25']]
```

```python
FE_data_needed
```

```python
FE_data_needed.to_csv('Final_fixed_data_needed.csv')
```

## Random Effect Model with Different Variables from Fixed Effect Model

```python
# Random Effect
y_var_name = 'mean_rent_per'
X_var_names = ['property_sqft','current_rent','property_value', 'income_tax_rate',
              'corporate_tax_rate', 'state_sales_tax_rate',
              'max_local_sales_tax', 'avg_household_income', 'population',
       'white_pop', 'black_pop', 'asian_pop', 'pop_0_19', 'pop_20_29', 'pop_30_39','house_supply', 'crime_rate','Tbond5',
              'hist_risk','SP500','starbucks_vists_per', 'colleges_per','public_schools_per', 'private_schools_per']
```

```python
from linearmodels import PanelOLS
from linearmodels import RandomEffects
exog_vars = X_var_names
exog = sm.add_constant(df[exog_vars])
model_re = RandomEffects(df[y_var_name], exog, check_rank=False) 
re_res = model_re.fit()
print(re_res)
```
